```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df= pd.read_csv(r"C:\Users\Admin\Desktop\DS\Projects\ML\homeprice_multiple.csv", encoding= 'unicode_escape')
```

```python
df.head()
```

```python
df = df.rename (columns= {"are":'area'})
```

```python
import math
b=math.floor(df.bedroom.median())
print(b)
```

```python
df.bedroom = df.bedroom.fillna(b)
```

```python
df.head()
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area', 'bedroom', 'age']], df.price)
```
